###### Department wise codekata performence (pie chart)

```python
Dept_Codekata_Perf = df.groupby('Department').sum()
Dept_Codekata_Perf
```

```python
Codekata_Perf = list(Dept_Codekata_Perf['CodeKata Score'])
Codekata_Perf
```

```python
Departments = list(Dept_Codekata_Perf.index)
Departments
```

```python
values = Codekata_Perf
labels = Departments
plt.figure(figsize=(6,10))
textprops = {"fontsize":15}
plt.pie(values,shadow = True,labels = labels,colors = ["blue","red","yellow"],autopct = "%2.2f%%",textprops =textprops)
plt.title("Dept_Wise Codekata Performance Piechart",fontsize = 25)
plt.legend(title = "Departments",loc=3, bbox_to_anchor=(1,0.5,1,1),fontsize=12)
plt.show()
```

###### Department wise toppers (horizantal bar graph or any visual representations of your choice)

```python
Toppers = df.groupby(['Department']).apply(lambda gdf:gdf.assign(Max_Score=lambda df: df['CodeKata Score'].max()))
Toppers
```

```python
Dept_Toppers = Toppers.loc[Toppers['CodeKata Score'] == Toppers['Max_Score']]
Dept_Toppers
```

```python
x = list(Dept_Toppers['Department'])
x
```

```python
y = list(Dept_Toppers['CodeKata Score'])
y
```

```python
data = list(Dept_Toppers['Name'])
data
```

```python
plt.barh(x,y, height = 0.5, color = ['green','red','brown'],edgecolor = 'black')
plt.xlabel("Codekata_Score")
plt.ylabel("Departments")
plt.title("Department Wise Toppers")
for index, value in enumerate(x):
  plt.text(index+y[index],value,data[index])
```
